```python
from sklearn import datasets
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = datasets.load_iris()
```

```python
print(iris.keys())
```

```python
x = pd.DataFrame(iris['data'], columns=iris['feature_names'])
y = pd.DataFrame(iris['target'], columns=['target_names'])
```

```python
data = pd.concat([x,y], axis=1)
data.head(5)
```

```python
# one_hot = pd.get_dummies(data['target_names'], prefix='target ')
# data = data.drop('target_names',1)
# data = pd.concat([one_hot, data], axis=1)
# data
```

```python
data.plot.scatter(x='sepal length (cm)', y='sepal width (cm)', c='target_names')
```

```python
color = {0:'r', 1:'g', 2:'b'}
data['color'] = data['target_names'].map(color)
data.head()
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-dark')
```

```python
data.plot.scatter(x='sepal length (cm)', y='sepal width (cm)', c=data['color'])
```

```python
target_names = {
    0:'setosa',
    1:'versicolor',
    2:'virginica'
}

data['target_names'] = data['target_names'].map(target_names)
```

```python
sns.lmplot('sepal length (cm)', 'sepal width (cm)', data=data, fit_reg=True, hue='target_names')
```

```python
sns.lmplot('sepal length (cm)', 'sepal width (cm)', data=data, fit_reg=False, hue='target_names')
```
